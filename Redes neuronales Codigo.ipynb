```python
#Se define la clase
class Perception:
  # Constructor: Se llama al crear una nueva instancia del Perceptrón.
  # Lista de modelos, inicializa los pesos, los sesgos, y la Taza de aprendizaje.
  def __init__(self, pesos, sesgo, taza_de_aprendisaje):
    self. pesos = pesos
    self. sesgo = sesgo
    self. taza_de_aprendizaje = taza_de_aprendisaje

  # Definimos la Función de Activación.
  # Convierte la entrada ponderada 'z' en una salida binaria (0 o 1).
  def funcion_de_activacion(self, z):
    if z>= 0:
        return 1
    else:
        return 0

  def entrenamiento(self, x, y, epocas):
    for epoca in range(epocas):
      for dato in x:
        x1, x2 = dato[0], dato[1]
        w1 = self.pesos[0]
        w2 = self.pesos[1]
        b = self.sesgo
        z = x1 * w1 + x2 * w2 + b
        y_pred = self.funcion_de_activacion(z)
        i = x.index(dato)
        y_real = y[i]
        self.pesos[0] = w1 + self.taza_de_aprendizaje * (y_real - y_pred) * x1
        self.pesos[1] = w2 + self.taza_de_aprendizaje * (y_real - y_pred) * x2
        self.sesgo = b + self.taza_de_aprendizaje * (y_real - y_pred)
        print (f"Dato{dato}, Prediction {y_pred}, Pesos {self.pesos}, Sesgo {self.sesgo},")
        print (f" Fin de la epoca {epoca}")
```

```python
x=[[1,1], [1,0], [0,1], [0,0]]
y=[1,0,0,0]
pesos=[2,3]
sesgo=4
taza_de_aprendizaje=0.1
```

```python
p = Perception(pesos, sesgo, taza_de_aprendizaje)
```

```python
p.entrenamiento(x,y,20)
```
